```python
import requests
import pandas
import boto3


def alphavantage(data, ticker):
    """ Take stock data from Alpha Vantage and pull into a csv file """

    # All links to time series data from Alpha Vantage
    url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=" + ticker + "&outputsize=full&apikey=YJH7Y8G3PCSKU5WP"


    # Request the data from the API URL provided
    data = requests.get(url)
    # Turn into JSON Data
    json_data = data.json()

    # Create an empty list for the dictionary keys
    keys = []
    # Function adding the dictionary keys into the list
    for key in json_data.keys():
        keys.append(key)

    # subset the JSON data into two data sets: meta data and the actual time series data
    meta_data = json_data[keys[0]]
    time_series = json_data[keys[1]]


    

    return time_series
    
```

```python
data = alphavantage(data="time series daily", ticker="MSFT")
```

```python
 time_series_df = pandas.DataFrame.from_dict(data, orient='index')
```

```python
time_series_df.reset_index(inplace=True)
```
